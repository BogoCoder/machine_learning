```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Creamos datos basados en una función
x = np.linspace(-5,5,100).reshape(-1,1)
y = x + np.sin(x)

yn = x + np.sin(x) + 1.5*np.random.randn(100,1)
```

```python
# Visualizamos
plt.plot(x, y, 'r', label = "original");
plt.scatter(x, yn, label = "random");
plt.xlabel("$x$");
plt.ylabel("$y$");
plt.legend()
```

```python
# Calculamos solución analítica
A = np.column_stack((np.ones((len(x),1)),x))
thetas = np.linalg.inv(A.T.dot(A)).dot(A.T).dot(yn)
print(thetas)
```

```python
h = np.matmul(A,thetas)
```

```python
plt.plot(x, y, 'r', label = "original");
plt.scatter(x, yn, label = "random");
plt.plot(x,h,'black', label = "regresión");
plt.xlabel('$x$');
plt.ylabel('$y$');
plt.legend();
```

Ya visualizamos nuestros datos, podemos proceder a la implementación del gradiente descendiente y comparar con la solución canónica:

```python
# Cálculo del costo
def cost(theta, x, y):
    
    h = x.dot(theta)
    m = len(y)
    cost = (1/2*m) * np.sum(np.square(h-y))
    return cost
```

```python
# Función de gradiente descendiente, iteraciones fijas en 100, puede cambiarla modificando la variable 'iter'
def grad_desc(x, y, theta, lrate=0.01):

    iter = 100
    
    x_A = np.c_[np.ones((len(x),1)),x]
    
    m = len(y)
    costs = np.zeros(iter)
    
    for it in range(iter):
        h = x_A.dot(theta)
        theta = theta - lrate*(1/m)*(x_A.T.dot(h-y))
        costs[it]  = cost(theta,x_A,y)
        
    return theta, costs
```

```python
# Función para generar interactividad
from ipywidgets import *

def auto(lrate):
    
    # Verificamos dimensionalidad y extraemos
    if len(x.shape) >= 2:
        num_rows, num_cols = x.shape
    else:
        num_cols = 1
        
    # Generamos un theta aleatorio y corremos la función grad_desc, print del costo final y thetas finales
    lrate = float(lrate)
    theta = np.random.randn(num_cols+1,1)
    theta,costs = grad_desc(x,y,theta,lrate)
    
    print("costo final: ",costs[-1])
    print("thetas: \n", theta )
    
    # Graficamos los costos en cada iteración, así como la regresión lineal generada por los thetas encontrados
    fig,(ax1, ax2) = plt.subplots(1,2,figsize=(16,8))

    iter = 100
    ax1.plot(range(iter),costs,'b.')
    ax1.set_ylabel('$J(\Theta)$')
    ax1.set_xlabel('Iteración')
    
    A = np.column_stack((np.ones((100,1)),x))
    h = np.matmul(A,theta)
    ax2.plot(x, y, 'r', label = "original");
    ax2.scatter(x, yn, label = "random");
    ax2.plot(x,h,'black', label = "regresión")
    ax2.set(xlabel = "$x$", ylabel = "$y$")
    ax2.legend()
```

Ejecutando nuestro programa interactivo:

```python
# Para cambiar el learning rate modifique la caja 'lrate'
# Cada ejecución comienza en un theta distinto aleatorio
widgets.interact_manual(auto, lrate = '0.01');
```

```python
plt.plot(x1, y, 'r', label = "original");
plt.scatter(x1, yn, label = "random");
plt.plot(x1,h,'black', label = "regresión");
plt.xlabel('$x_1$');
plt.ylabel('$y$');
plt.legend();
```

```python
# Función para generar interactividad
from ipywidgets import *

def auto(lrate):
    
    # Verificamos dimensionalidad y extraemos
    if len(x.shape) >= 2:
        num_rows, num_cols = x.shape
    else:
        num_cols = 1
        
    # Generamos un theta aleatorio y corremos la función grad_desc, print del costo final y thetas finales
    lrate = float(lrate)
    theta = np.random.randn(num_cols+1,1)
    theta,costs = grad_desc(x,y,theta,lrate)
    
    print("costo final: ",costs[-1])
    print("thetas: \n", theta )
    
    # Graficamos los costos en cada iteración, así como la regresión lineal generada por los thetas encontrados
    fig,(ax1, ax2) = plt.subplots(1,2,figsize=(16,8))

    iter = 100
    ax1.plot(range(iter),costs,'b.')
    ax1.set_ylabel('$J(\Theta)$')
    ax1.set_xlabel('Iteración')
    
    A = np.column_stack((np.ones((100,1)),x))
    h = np.matmul(A,theta)
    ax2.plot(x1, y, 'r', label = "original");
    ax2.scatter(x1, yn, label = "random");
    ax2.plot(x1,h,'black', label = "regresión")
    ax2.set(xlabel = "$x_1$", ylabel = "$y$")
    ax2.legend()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Cargando el archivo de excel con los datos
data = pd.read_excel('Real estate valuation data set.xlsx')
```

```python
# Regresión lineal simple

x = data['X2 house age']
y = data['Y house price of unit area']

A = np.column_stack((np.ones((len(x),1)),x))
thetas = np.linalg.inv(A.T.dot(A)).dot(A.T).dot(y)

print(thetas)
```

```python
h_1 = np.matmul(A,thetas)
```

```python
# Visualizamos alguna variable de interés
plt.scatter(data['X2 house age'], data['Y house price of unit area'])
plt.plot(data['X2 house age'],h_1,c='r')
plt.xlabel('Age [years]')
plt.ylabel('Price [dollars/$m^2$]')
```

```python
# Regresión lineal multivariable

x = data[['X2 house age', 'X3 distance to the nearest MRT station', 'X4 number of convenience stores']]
y = data['Y house price of unit area']

A = np.column_stack((np.ones((len(x),1)),x))
thetas = np.linalg.inv(A.T.dot(A)).dot(A.T).dot(y)

print(thetas)
```

```python
h_2 = np.matmul(A,thetas)
```

```python
# Visualizamos alguna variable de interés
plt.scatter(data['X3 distance to the nearest MRT station'], data['Y house price of unit area'])
plt.scatter(x['X3 distance to the nearest MRT station'],h_2,c='r')
plt.xlabel('Distance')
plt.ylabel('Price [dollars/$m^2$]')
```

```python
# Evaluamos los errores y comparamos
e_1 = np.matmul(np.transpose((h_1-y)),(h_1-y))/len(x)
e_2 = np.matmul(np.transpose((h_2-y)),(h_2-y))/len(x)
print("Error cuadrático medio del modelo lineal :",e_1)
print("Error cuadrático medio del modelo lineal multivariable :",e_2)
```

Evaluando los errores para los modelos construidos, podemos inferir que el mejor modelo es el multivariable, pues su error es menor, aunque hay que tener cuidado con la dimensionalidad, pues podría ocasionar overfitting.
